### 3. Finding Missing Values
##### Task 2: Identify missing values for all columns.


```python
## Write your code here
missing_data = df.isna().sum()
print(missing_data)
```

##### Task 3: Visualize missing values using a heatmap (Using seaborn library).



```python
## Write your code here
import seaborn as sns
import matplotlib.pyplot as plt

# Visualize missing values
plt.figure(figsize=(12,6))
sns.heatmap(df.isna(), 
            cmap="viridis",   # color scheme
            cbar=False)       # remove color bar
plt.title("Heatmap of Missing Values", fontsize=16)
plt.show()
```

##### Task 4: Count the number of missing rows for a specific column (e.g., `Employment`).


```python
## Write your code here
df['Employment'].isna().sum()
```

### 4. Imputing Missing Values
##### Task 5: Identify the most frequent (majority) value in a specific column (e.g., `Employment`).


```python
## Write your code here
most_frequent_employment = df['Employment'].mode()[0]
print(f'Most frequent employment is: {most_frequent_employment}')
```

##### Task 6: Impute missing values in the `Employment` column with the most frequent value.



```python
## NB there is no missing data in column employment, we can see in the code above
import numpy as np
df['Employment'].replace(np.nan, 'Employed, full time')
```

### 5. Visualizing Imputed Data
##### Task 7: Visualize the distribution of a column after imputation (e.g., `Employment`).


```python
top_employment = df['Employment'].value_counts().nlargest(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_employment.values, y=top_employment.index, palette="viridis")

plt.title("Top 10 Employment Categories after Imputation")
plt.xlabel("Count")
plt.ylabel("Employment Category")
plt.show()
```

```python
df['Employment'].value_counts().nlargest(8).plot.pie(
    autopct='%1.1f%%', figsize=(7,7), colormap="Set3"
)
plt.title("Employment Distribution (Top 8 categories)")
plt.ylabel("")
plt.show()
```
